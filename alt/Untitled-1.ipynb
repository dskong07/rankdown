```python
import pandas as pd
import requests

def format_df(df, name_col, value_col, metric):
    df = df[[name_col, value_col]].copy()
    df.columns = ["name", "value"]
    df = df.sort_values("value", ascending=False).reset_index(drop=True)
    df["rank"] = df.index + 1
    df["metric"] = metric
    return df[["rank", "name", "metric", "value"]].head(200)

# ----------------------------
# 1. Population (REAL)
# ----------------------------
pop_url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
pop = pd.read_csv(pop_url)
pop = pop[pop["Year"] == pop["Year"].max()]

pop_df = format_df(pop, "Country Name", "Value", "population")
pop_df.to_csv("countries_population_top200.csv", index=False)



print("Population + Land Area CSVs generated")

# ----------------------------
# 3. Market Cap (ADD FILE)
# ----------------------------
# Download CSV from:
# https://companiesmarketcap.com/
# Save as: companies_market_cap_2025.csv

market = pd.read_csv("companies_mkt_cap.csv")
mkt_df = format_df(market, "Company", "MarketCap", "market_cap")
mkt_df.to_csv("companies_market_cap_2025_top200.csv", index=False)

# # ----------------------------
# # 4. Actors (ADD FILE)
# # ----------------------------
# # Use IMDb datasets:
# # https://datasets.imdbws.com/
# # Filter titleType == movie, count per actor
# # Save as: actors_film_counts.csv

# actors = pd.read_csv("actors_film_counts.csv")
# act_df = format_df(actors, "Actor", "FilmCount", "film_appearances")
# act_df.to_csv("actors_most_movies_top200.csv", index=False)

# # ----------------------------
# # 5. Peak Temperature (ADD FILE)
# # ----------------------------
# # Use curated dataset (WMO verified)
# # Save as: country_peak_temperatures.csv

# temps = pd.read_csv("country_peak_temperatures.csv")
# temp_df = format_df(temps, "Country", "TemperatureC", "peak_temperature")
# temp_df.to_csv("countries_peak_temperature_top200.csv", index=False)

# print("All CSVs generated.")
```
